```python
import numpy as np
```

```python
# 시그모이드 함수
def actf(x):
    print("x in actf=", x)
    return 1/(1+np.exp(-x))
```

```python
# 시그모이드 함수의 미분치
def actf_deriv(x):
    return x*(1-x)
```

```python
# 입력유닛의 개수, 은닉유닛의 개수, 출력유닛의 개수
inputs, hiddens, outputs = 2, 2, 1
learning_rate=0.2
```

```python
# 훈련 샘플과 정답
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
T = np.array([[0], [1], [1], [0]])
```

```python
X
```

```python
W1 = np.array([[0.10, 0.20],
                 [0.30, 0.40]])
W2 = np.array([[0.50], [0.60]])
B1 = np.array([0.1, 0.2])
B2 = np.array([0.3])
```

```python
W1
```

```python
# 순방향 전파 계산
def predict(x):
    layer0 = x                   # 입력을 layer0에 대입한다.
    print("layer0=", layer0)
    Z1 = np.dot(layer0, W1)+B1   # 행렬의 곱을 계산한다.
    print("Z1=", Z1)
    layer1 = actf(Z1)            # 활성화 함수를 적용한다.
    print("layer1=", layer1, "\n")
    Z2 = np.dot(layer1, W2)+B2   # 행렬의 곱을 계산한다.
    print("Z2=", Z2)
    layer2 = actf(Z2)            # 활성화 함수를 적용한다.
    print("layer2=", layer2, "\n")
    return layer0, layer1, layer2
```

```python
def test():
    for x, y in zip(X, T):
        x = np.reshape(x, (1, -1))    # x를 2차원 행렬로 만든다.입력은 2차원이어야 한다.
        layer0, layer1, layer2 = predict(x)
        print(x, y, layer2)
```

```python
test()
```
